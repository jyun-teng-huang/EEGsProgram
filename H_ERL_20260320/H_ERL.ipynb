```python
import numpy as np

def euclidean_distance(E1, E2):
    """
    計算兩個協方差矩陣之間的歐氏距離 (Frobenius Norm)
    """
    # 直接計算矩陣差的 Frobenius 範數
    return np.linalg.norm(E1 - E2)

def riemannian_distance(E1, E2):
    """
    計算兩個協方差矩陣之間的黎曼距離 (Affine-Invariant Riemannian Metric)
    """
    # 根據公式 d = sqrt(sum(log(eig(inv(E1) * E2))^2))
    # 為了數值穩定性，通常建議使用廣義特徵值，但在只用 numpy 的情況下：

    # 1. 計算 E1 的逆矩陣
    E1_inv = np.linalg.inv(E1)

    # 2. 計算 inv(E1) * E2 的特徵值
    # 注意：雖然結果理論上是實數，但數值計算可能產生微小複數，取實部
    eigvals = np.linalg.eigvals(np.dot(E1_inv, E2))

    # 3. 計算對數平方和再開根號
    # 加上一個極小值 epsilon 防止 log(0)
    dist = np.sqrt(np.sum(np.log(np.real(eigvals))**2))

    return dist

# --- 測試範例 ---
if __name__ == "__main__":
    # 建立兩個 3x3 的對稱正定矩陣 (模擬 EEG 3 通道的 Covariance)
    A = np.array([[1.2, 0.3, 0.1],
                  [0.3, 1.0, 0.2],
                  [0.1, 0.2, 0.8]])

    B = np.array([[1.5, 0.2, 0.2],
                  [0.2, 1.3, 0.4],
                  [0.2, 0.4, 1.1]])

    d_e = euclidean_distance(A, B)
    d_r = riemannian_distance(A, B)

    print(f"歐氏距離 (Euclidean): {d_e:.6f}")
    print(f"黎曼距離 (Riemannian): {d_r:.6f}")
```
